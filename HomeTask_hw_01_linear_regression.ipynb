## Задание 5

Подберите оптимальный градиентный шаг `eta0` для SGDRegressor. Перебирайте шаг в диапазоне от 0.00001 до 0.1 не включительно с шагом 0.0001.

Какого наилучшего качества на тесте удалось добиться? Полученное значение $R^2$ округлите до сотых.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# your code here

eta0_dict = dict()

for i in np.arange(0.00001, 0.1, 0.0001):

    reg_e0 = SGDRegressor(eta0=i)

    reg_e0.fit(X_train, y_train)

    pred_e0 = reg_e0.predict(X_test)

    r2 = r2_score(y_test,pred_e0)

    eta0_dict[i] = r2

    print(f'eta0:{round(i, 5)} -, R2: {round(r2, 3)}')

# Ключ, значение которого является наибольшим
key_max = max(eta0_dict, key = lambda k: eta0_dict[k])

print("The key with the largest value:", round(key_max, 5))

r2_eta =  eta0_dict[key_max]
# Получение наибольшего значения
print("The largest value:", round(r2_eta, 2))
```

```python
# можно конечно искать махимум на лету, без словаря (но даже в этом примере я получил минимум два схожих результата и если не записывать в словрь то разобраться ->...)
# во всяком случае гоу

eta = 0
r2 = 0

for e in np.arange(.00001, .1, .0001):

  sgd = SGDRegressor(eta0=e)
  sgd.fit(X_train, y_train)

  r2_i = r2_score(y_test, sgd.predict(X_test))

  if r2_i > r2:
    eta = e
    r2 = r2_i

print(f"eta0={eta}: R2={round(r2, 2)}")
```
